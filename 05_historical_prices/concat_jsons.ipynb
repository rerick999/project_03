```python
import pandas as pd
from datetime import timedelta
import datetime as dt
import json
import os
import glob
import numpy as np
```

```python
# declare source directory/paths
current_dir = os.getcwd()
parent_dir = os.path.dirname(current_dir)
prices_dir = os.path.join(parent_dir,'stock_scrape_load','data')

prices_path = f"{prices_dir}/*.json"
names_path = os.path.join(parent_dir,'05_michael_test/company_info.csv')
history_path = os.path.join(parent_dir,'05_michael_test','stock_data2.json')
```

```python
# declare empty list to store data
data = []
past_history = []

# Loop through all json files and update data list
for file in glob.glob(prices_path):
    with open(file, 'r') as json_prices:
        json_prices_data = json.load(json_prices)
        data.append(json_prices_data)

for file in glob.glob(history_path):
    with open(file, 'r') as json_history:
        json_history_data = json.load(json_history)
        past_history.append(json_history_data)

# Create company name / ticker symbol Dataframe
names_df = pd.read_csv(names_path)
names_df = names_df[['ticker','shortName']]

# Create hitorical prices Dataframe
prices_df = pd.concat([pd.DataFrame(d) for d in data], ignore_index = True)
prices_df['date'] = pd.to_datetime(prices_df['date'])
prices_df = prices_df.replace(np.nan, None)
past_history = pd.DataFrame(pd.read_json(history_path)).drop(columns='Adj Close')
past_history = pd.merge(past_history,names_df,how = 'left', on = 'ticker')



# View that sweet sweet extended dataframe!
# prices_df = prices_df.sort_values(['ticker','date'])
past_history.head(20)
```

```python
historical_json = []
grouped_tickers = past_history.groupby(['ticker'])
grouped_dates = past_history.groupby(['Date'])
grouped_companies = past_history.groupby(['shortName'])
data = {
    "tickers": list(grouped_tickers.groups.keys()),
    "companies":list(grouped_companies.groups.keys()),
    "dates": list(grouped_dates.groups.keys()),
    "history": []
}

for ticker, group in grouped_tickers:
    for index, row in group.iterrows():
        ticker_data = {
            "ticker": ticker,
            "name": row['shortName'],
            "date": row['Date'],
            "values": [{
                "date": row['Date'],
                "previousClose": row['Close'],
                "dailyOpen": row['Open'],
                "dayHigh": row['High'],
                "dayLow": row['Low'],
                # "ask": row['ask'],
                # "bid": row['bid'],
                "volume": row['Volume']
            }]
        }
        data["history"].append(ticker_data)

# json cannot work with datetime dtypes: https://www.geeksforgeeks.org/how-to-fix-datetime-datetime-not-json-serializable-in-python/
def serialize_dt(obj):
    if isinstance(obj, dt.datetime):
        return obj.isoformat()
    # elif isinstance(obj, float) and np.isnan(obj):
    #     return None  # Convert NaN to null
    raise TypeError("Type not serializable")

# Write json to 'visualizations' directory
json_data = json.dumps(data, indent=4, default = serialize_dt) #, cls = CustomEncoder)
with open("../visualizations/historical.json", "w") as json_file:
    json_file.write(json_data)
```
