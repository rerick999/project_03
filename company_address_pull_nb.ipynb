```python
import yfinance as yf
import csv
import os
import json

current_directory = os.getcwd()


msft = yf.Ticker("MSFT")
```

```python
import pprint
pprint.pprint(msft.info['zip'])
```

```python
def get_data(ticker):
    print("processing... ", ticker)
    company = yf.Ticker(ticker)
    address = company.info['address1']
    city = company.info['city']
    state = company.info['state']
    zip = company.info['zip']
    data = {ticker:{
        'address':address,
        'city':city,
        'state':state,
        'zip':zip
    }}
    return(data)
```

```python
filename = "top_100_company_tickers.csv"

company_data = {}

with open(filename, 'r') as myfile:
    reader = csv.DictReader(myfile)
    for row in reader:
        ticker = row['Ticker']
        try:
            row_data = get_data(ticker)
            company_data.update(row_data)
        except Exception as e:
            print("error processing ", ticker + ":", e)
```

```python
company_data
```

```python
output_filename = "company_data.json"
with open(output_filename, 'w') as outfile:
    json.dump(company_data, outfile)

print("Company data saved as...", output_filename)
```

```python
fieldnames = ["ticker", "address", "city", "state", "zip"]

output_filename = "company_data.csv"
output_filepath = os.path.join(current_directory, output_filename)

with open(output_filename, 'w', newline='') as csvfile:
    writer = csv.DictWriter(csvfile, fieldnames=fieldnames)
    writer.writeheader()
    for ticker, data in company_data.items():
        writer.writerow({"ticker": ticker, "address": data.get("address", ""),
                         "city": data.get("city", ""), "state": data.get("state", ""),
                         "zip": data.get("zip", "")})

print("Company data saved as", output_filename)
```
